```python
import csv
import os

root = os.getcwd()

honda_folder = os.path.join(root, 'Honda')
yamaha_folder = os.path.join(root, 'Yamaha')
suzuki_folder = os.path.join(root, 'Suzuki')
vinfast_folder = os.path.join(root, 'Vinfast')
others_folder = os.path.join(root, 'Others')
```

```python
image_counts_by_student = {}

image_counts_by_brand_from_student = {}

dataset = [honda_folder, yamaha_folder , suzuki_folder, vinfast_folder, others_folder]

for brand_folder in dataset:
    for img_path in os.listdir(brand_folder):
        filename = os.path.basename(img_path)
        parts = filename.split('.')

        students_id = parts[0].split('-')
        if students_id == 215202569:
          students_id = 21520569
        brand = parts[1]

        for id in students_id:

            if id not in image_counts_by_student:
              image_counts_by_student[id] = 0
            image_counts_by_student[id] += 1.0 / len(students_id)

            if id not in image_counts_by_brand_from_student:
              image_counts_by_brand_from_student[id] = {}

            if brand not in image_counts_by_brand_from_student[id]:
              image_counts_by_brand_from_student[id][brand] = 0
            image_counts_by_brand_from_student[id][brand] += 1.0 / len(students_id)
```

```python
cd /content/drive/MyDrive/
```

```python
# Ghi thông tin vào file MotocycleDataset-1.csv
with open("MotocycleDataset-1.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["MSSV", "All", "Số lượng"])
    for student_id, count in image_counts_by_student.items():
        writer.writerow([student_id, "All", round(count, 2)])

# Ghi thông tin vào file MotocycleDataset-2.csv
with open("MotocycleDataset-2.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["MSSV", "Hiệu xe", "Số lượng"])
    for student_id,brand_and_count in image_counts_by_brand_from_student.items():
      for brand,count in brand_and_count.items():
        writer.writerow([student_id, brand, round(count, 2)])
```
